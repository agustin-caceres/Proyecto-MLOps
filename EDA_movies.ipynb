### **Distribución del Promedio de Votos**


El gráfico muestra la distribución de las calificaciones promedio (votos) de las películas en el dataset. Se puede observar lo siguiente:

1. **Concentración de votos**: La mayoría de las películas tienden a obtener una calificación promedio entre 5 y 7, con un pico claro alrededor de 6, lo que indica que muchas películas reciben una valoración moderada por parte de la audiencia.
   
2. **Calificaciones extremas**: Existen pocas películas con calificaciones muy bajas (por debajo de 2) o muy altas (por encima de 8). Esto sugiere que ambos extremos (excelentes o muy malas) son menos comunes.

- **Conclusión**: La mayoría de las películas obtienen una calificación promedio moderada por parte de la audiencia, con pocas alcanzando calificaciones extremadamente bajas o altas.




```python
import matplotlib.pyplot as plt

# Se crea la figura y los ejes
plt.figure(figsize=(10,6))

# Histograma de 'vote_average'
plt.hist(movies_df['vote_average'], bins=40, edgecolor='black', linewidth= 1.2)

# Título y etiquetas
plt.title('Distribución del Promedio de Votos', fontsize=16)
plt.xlabel('Promedio de Votos', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)

plt.show()

```

### **Distribución de peliculas a lo largo del tiempo**

Este gráfico de barras muestra la cantidad de películas lanzadas por año a lo largo del tiempo.

1. **Tendencia creciente**: Se puede notar que a partir de la década de 1990 en adelante, hubo un notable aumento en la cantidad de películas producidas. Este crecimiento puede deberse a varios factores, como el avance de la tecnología, la expansión del cine a nivel global y el crecimiento de plataformas de distribución digital.

2. **Estancamiento hasta 1960**: Desde los inicios del cine hasta la década de 1960, la cantidad de películas producidas por año fue relativamente baja. Esto se debe en parte a las limitaciones tecnológicas, económicas y de distribución de aquella época.

3. **Explosión en los 2000s**: A partir de los años 2000, el número de películas creció de manera exponencial, alcanzando su punto máximo en los últimos años. Este auge podría estar relacionado con la evolución de las plataformas de streaming, la accesibilidad de las herramientas de producción y la globalización del mercado del entretenimiento.



```python
import seaborn as sns

# figura y los ejes
plt.figure(figsize=(24,8))

# Gráfico de barras de 'release_year'
sns.countplot(data=movies_df, x='release_year', palette='viridis')

# Se rotan etiquetas en el eje X 
plt.xticks(rotation=90)

plt.title('Cantidad de Películas por Año', fontsize=16)
plt.xlabel('Año de Lanzamiento', fontsize=12)
plt.ylabel('Cantidad', fontsize=12)

plt.show()
```

### **Distribución de la Duración de las Películas**

1. **Concentración de Duraciones**: La mayoría de las películas tienen una duración entre **80 y 150 minutos**. El pico más pronunciado se encuentra alrededor de los **100 minutos**, por ende esta es la duración promedio más común para las películas.

2. **Casos Atípicos**: Existen algunas películas con duraciones mayores a **200 minutos** (más de 3 horas). También se observan algunas películas con duraciones inferiores a **60 minutos**. Ambos casos son menos comunes comparadas con las de duración estándar.

- **Conclusión**: La mayoría de las películas siguen un patrón de duración estándar que va de **75 a 150 minutos**. Esto refleja que las compañías prefieren mantener las películas en un rango que resulte cómodo para los espectadores.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# figura y los ejes
plt.figure(figsize=(10, 6))

# Histograma de 'runtime'
sns.histplot(movies_df['runtime'].dropna(), bins=30, color='#01377D', edgecolor='black')

# Se ajustan los valores extremos
plt.xlim(0, 300)

# Título y etiquetas
plt.title('Distribución de la Duración de las Películas', fontsize=16)
plt.xlabel('Duración (min)', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)

plt.show()
```

### **Distribución de Películas por Idioma (Top 10)**

El gráfico de barras muestra un **Top Ten** de los idiomas más comunes en los que se producen las películas en el dataset.

1. **Dominio del Inglés**: Es el idioma predominante, con una diferencia abrumadora respecto a los demás idiomas. Esto refleja la influencia de la industria de cine en países de habla inglesa, principalmente Hollywood.

2. **Idiomas Europeos y Asiáticos**: Otros idiomas que destacan en la lista son el **francés (fr)**, **alemán (de)**, **italiano (it)**, y **español (es)**, que reflejan la producción cinematográfica en Europa, así como el **japonés (ja)** y **chino (zh)** que representan la producción en Asia.

3. **Idiomas Escandinavos y Eslavos**: También aparecen algunos idiomas menos hablados a nivel mundial, como el **sueco (sv)**, el **ruso (ru)**, y el **danés (da)**, que representan cinematografías más pequeñas.

- **Conclusión**: El inglés domina claramente la producción cinematográfica, pero se observa una diversidad interesante en las películas producidas en otros idiomas, lo que refleja el alcance global de la industria del cine.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Se crea la figura
plt.figure(figsize=(12, 5))

# Gráfico de barras de 'original_language'
sns.countplot(
    y='original_language', 
    data=movies_df, 
    order=movies_df['original_language'].value_counts().index[:10], 
    palette='Blues_r'
)

# Se agregan los valores exactos en las barras
for index, value in enumerate(movies_df['original_language'].value_counts().values[:10]):
    plt.text(value + 350, index, str(value), color='black', va="center", ha="left", fontsize=10)

# Título y etiquetas
plt.title('Distribución de Películas por Idioma (Top 10)', fontsize=16)
plt.xlabel('Cantidad de Películas', fontsize=12)
plt.ylabel('Idioma Original', fontsize=12)

# Se mejora el diseño general del gráfico
plt.tight_layout()
plt.show()
```

### **Distribución de películas por Género**

Se puede notar que el género **"Drama"** está presente en la gran mayoría de producciones, pero también se observa una gran variedad.

```python
# Se cuenta la frecuencia de cada valor en la columna 'genre_names'
genre_counts = movies_df['genre_name'].value_counts()

# Se eliminan los géneros que solo aparecen una vez
genre_counts = genre_counts[genre_counts > 1]

# Se crea el gráfico de barras
plt.figure(figsize=(15, 6))
sns.barplot(x=genre_counts.index, y=genre_counts.values, palette='magma', saturation= 0.7)

# Rotación de etiquetas en el eje X con un ángulo de 30 grados
plt.xticks(rotation=40, ha='right')

# Título y etiquetas
plt.title('Número de Películas por Género', fontsize=16)
plt.xlabel('Género', fontsize=12)
plt.ylabel('Cantidad de Películas', fontsize=12)

# Se muestra el gráfico con mejor ajuste de márgenes
plt.tight_layout()
plt.show()
```

### **Relación entre el Presupuesto y los Ingresos**

Este gráfico de dispersión muestra la relación entre el presupuesto de las películas (eje X) y sus ingresos (eje Y) destacando las siguientes conclusiones:

1. **Concentración de películas con presupuestos bajos (-100 millones):** 
   - La mayoría de las películas tienen un presupuesto que se encuentra por debajo de los **100 millones de dólares** (1e8), lo que significa que las grandes producciones son la excepción.

2. **Ingresos dispersos en películas de bajo presupuesto:**
   - Aunque muchas películas tienen presupuestos bajos (menos de 100 millones), sus ingresos pueden variar considerablemente, algunas llegando a generar ingresos superiores a los **500 millones de dólares (0.5e9)**.
   - Esto muestra que no siempre es necesario un gran presupuesto para generar ingresos significativos.

3. **Películas con presupuestos altos (+100 millones):**
   - Existen algunas películas con presupuestos superiores a los **100 millones de dólares**, llegando incluso a los **300 millones (3e8)**. Estas películas tienden a generar mayores ingresos, con algunas alcanzando los **1,000 millones de dólares (1e9)** o más.

4. **Correlación no lineal:**
   - Si bien existe cierta correlación positiva entre el presupuesto y los ingresos, la relación no es estrictamente lineal. Algunas películas de presupuesto moderado han generado ingresos muy altos, mientras que otras, con presupuestos elevados, no han tenido el mismo éxito en taquilla.

5. **Calificaciones de la audiencia (promedio de votos):**
   - La escala de colores muestra la calificación promedio (vote_average) de cada película. Se puede notar que películas con calificaciones promedio altas (de 8 a 10) están más dispersas en cuanto a presupuesto e ingresos. Sin embargo, hay muchas películas con bajos presupuestos y calificaciones moderadas que también han generado ingresos considerables.

- **En conclusión**, ninguna de las dos variables **(*Presupuesto* y *calificación de los votos*)** nos ofrece una relación clara y directa con los ingresos obtenidos por las películas.


```python
plt.figure(figsize=(10, 6))

sns.scatterplot(x='budget', y='revenue', hue='vote_average', size='vote_average',
                data=movies_df, palette='Blues', sizes=(40,120), alpha=0.7, edgecolor='white')

# Título y etiquetas
plt.title('Relación entre el Presupuesto y los Ingresos (con promedio de votos)', fontsize=16)
plt.xlabel('Presupuesto', fontsize=12)
plt.ylabel('Ingresos', fontsize=12)

plt.show()
```

### **Relación de variables**: `Presupuesto`, `Ingresos`, `Calificaciones`, `Duración`
Hay una marcada correlación entre el presupuesto y los ingresos (0.77), lo cual es algo esperado. Sin embargo, el promedio de votos no está fuertemente correlacionado con ninguna otra variable, lo que refuerza la conclusión de que las calificaciones de los usuarios no son un factor decisivo en los ingresos.

```python
numeric_columns = movies_df[['budget', 'revenue', 'vote_average', 'runtime']]

# Se obtiene la correlación
correlation_matrix = numeric_columns.corr()

# Visualización
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='RdBu_r', linewidths=0.5)
plt.title('Correlaciones de variabless')
plt.show()
```
